```python
import pandas as pd
vera = pd.read_csv("/incarceration_trends_county.csv", dtype=str)

print("Shape:", vera.shape)
print("Columns:", vera.columns.tolist()[:30])
print("\nSample rows:")
display(vera.head(10))

# check years
print("\nUnique years in dataset:")
print(vera['year'].unique()[:20])   # first 20 unique values

# count number of counties per year
coverage = vera.groupby('year')["fips"].nunique()
print("\nCounty coverage by year:")
print(coverage.sort_index())
```

```python
import pandas as pd

# filter to 2022 Q4 only
vera_2022_q4 = vera[(vera['year'] == "2022") & (vera['quarter'] == "4")].copy()

num_cols = vera_2022_q4.columns.difference(
    ["year", "quarter", "state_abbr", "state_code", "state_fips",
     "county_name", "county_code", "fips", "urbanicity", "region",
     "division", "metro_area_2000", "commuting_zone_2000"]
)
vera_2022_q4[num_cols] = vera_2022_q4[num_cols].apply(pd.to_numeric, errors="coerce")

# key vars
keep_cols = [
    "fips", "state_abbr", "county_name", "total_pop",
    "jail_total", "convicted_jail_pop", "unconvicted_jail_pop",
    "prison_pop", "jail_adm_tot"
]

vera_core = vera_2022_q4[keep_cols].copy()

# compute
vera_core["pretrial_rate_100k"] = (
    vera_core["unconvicted_jail_pop"] / vera_core["total_pop"]
) * 100000

vera_core["pretrial_share"] = (
    vera_core["unconvicted_jail_pop"] / vera_core["jail_total"]
)

print(vera_core.shape)
display(vera_core.head())
```
