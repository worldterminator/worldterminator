```python
import pandas as pd
import numpy as np

YEAR = "2022"
QUARTER = "2"
vera_2022_q2 = vera[(vera["year"] == YEAR) & (vera["quarter"] == QUARTER)].copy()

# id string
id_cols = [
    "year", "quarter", "state_abbr", "state_code", "state_fips",
    "county_name", "county_code", "fips", "urbanicity", "region",
    "division", "metro_area_2000", "commuting_zone_2000"
]
# standardize FIPS
if "fips" in vera_2022_q2.columns:
    vera_2022_q2["fips"] = vera_2022_q2["fips"].astype(str).str.zfill(5)

num_cols = [c for c in vera_2022_q2.columns if c not in id_cols]
vera_2022_q2[num_cols] = vera_2022_q2[num_cols].apply(pd.to_numeric, errors="coerce")
```

```python
# check coverage of pretrial count vs rate by quarter
v22 = vera[vera["year"] == "2022"].copy()
tbl = (v22
       .groupby("quarter")
       .agg(
           rows=("fips","count"),
           nonmissing_pretrial=("total_jail_pretrial", lambda s: int(s.notna().sum())),
           nonmissing_pretrial_rate=("total_jail_pretrial_rate", lambda s: int(s.notna().sum())),
       )
       .assign(
           pct_pretrial=lambda d: d["nonmissing_pretrial"]/d["rows"],
           pct_pretrial_rate=lambda d: d["nonmissing_pretrial_rate"]/d["rows"],
       )
       .sort_index())
print(tbl)
```

```python
# make new variables!
# total_pop as denominator as requested

den_col = "total_pop" if "total_pop" in vera_2022_q2.columns else None

def safe_div(n, d):
    n = pd.to_numeric(n, errors="coerce")
    d = pd.to_numeric(d, errors="coerce")
    return np.where((d.notna()) & (d != 0), n / d, np.nan)

# Rates won't error; rows with missing/zero denom become NaN
vera_2022_q2["pretrial_rate_100k"] = (
    safe_div(vera_2022_q2["total_jail_pretrial"], vera_2022_q2[den_col]) * 1e5
    if den_col else np.nan
)

vera_2022_q2["pretrial_share"] = safe_div(
    vera_2022_q2["total_jail_pretrial"], vera_2022_q2["total_jail_pop"]
)
# keep everything (all jail/prison/admissions/race fields) for merge later
vera_keep = vera_2022_q2.copy()
```

```python
import pandas as pd
enslaved = pd.read_csv("/content/nhgis0002_ds13_1860_county.csv")
totalpop = pd.read_csv("/content/nhgis0002_ds14_1860_county.csv")

enslaved = enslaved.iloc[1:].copy()
totalpop = totalpop.iloc[1:].copy()

# to numeric
enslaved["AGU001"] = pd.to_numeric(enslaved["AGU001"], errors="coerce")
totalpop["AG3001"] = pd.to_numeric (totalpop ["AG3001"], errors="coerce")
```

```python
# quick inspection
print("Enslaved 1860:")
print(enslaved.head(), "\n")
print(enslaved.info(), "\n")

print("Total Pop 1860:")
print(totalpop.head(), "\n")
print(totalpop.info(), "\n")
```

```python
enslaved = enslaved.rename(columns={"AGU001": "enslaved_1860"})
totalpop = totalpop.rename(columns={"AG3001": "total_pop_1860"})
```

```python
merged = (
    enslaved[["GISJOIN","STATE","COUNTY","STATEA","COUNTYA","enslaved_1860"]]
      .merge(totalpop[["GISJOIN","total_pop_1860"]], on="GISJOIN", how="outer")
)
```

```python
import numpy as np
merged["slave_share_1860"] = merged["enslaved_1860"] / merged["total_pop_1860"]
merged.loc[(merged["total_pop_1860"].fillna(0) == 0), "slave_share_1860"] = np.nan
```

```python
# derive 5-digit FIPS from STATEA+COUNTYA
# e.g. STATEA='010' -> '01'; COUNTYA='0010' -> '001'  => fips='01001'
merged["fips"] = merged["STATEA"].str[:2].fillna("") + merged["COUNTYA"].str[:3].fillna("")
merged["fips"] = merged["fips"].where(merged["fips"].str.len() == 5, np.nan)
```

```python
for col in ["STATE","COUNTY"]:
    if col in merged.columns:
        merged[col] = merged[col].str.upper().str.strip()
```

```python
print("Row counts â†’ enslaved:", len(enslaved), " total:", len(totalpop), " merged:", len(merged))
print("\nMissing after merge:\n", merged[["enslaved_1860","total_pop_1860","slave_share_1860","fips"]].isna().sum())
```

```python
# what didnâ€™t match
only_ensl = enslaved[~enslaved["GISJOIN"].isin(totalpop["GISJOIN"])][["GISJOIN","STATE","COUNTY"]]
only_tot  = totalpop[~totalpop["GISJOIN"].isin(enslaved["GISJOIN"])][["GISJOIN"]]
print("\nOnly in enslaved:", len(only_ensl))
print("Only in totalpop:", len(only_tot))
```

```python
print("\nRows only in enslaved:")
print(only_ensl)

print("\nRows only in totalpop:")
print(only_tot)
```

```python
# random spot chekcks
print(merged.sample(5, random_state=42)[["fips","STATE","COUNTY","enslaved_1860","total_pop_1860","slave_share_1860"]])

# highest and lowest slave shares
print(merged.sort_values("slave_share_1860", ascending=False)
             [["fips","STATE","COUNTY","enslaved_1860","total_pop_1860","slave_share_1860"]]
             .head(5))
```
