```python
import pandas as pd
import numpy as np

YEAR = "2022"
QUARTER = "2"
vera_2022_q2 = vera[(vera["year"] == YEAR) & (vera["quarter"] == QUARTER)].copy()

# id string
id_cols = [
    "year", "quarter", "state_abbr", "state_code", "state_fips",
    "county_name", "county_code", "fips", "urbanicity", "region",
    "division", "metro_area_2000", "commuting_zone_2000"
]
# standardize FIPS
if "fips" in vera_2022_q2.columns:
    vera_2022_q2["fips"] = vera_2022_q2["fips"].astype(str).str.zfill(5)

num_cols = [c for c in vera_2022_q2.columns if c not in id_cols]
vera_2022_q2[num_cols] = vera_2022_q2[num_cols].apply(pd.to_numeric, errors="coerce")
```

```python
# check coverage of pretrial count vs rate by quarter
v22 = vera[vera["year"] == "2022"].copy()
v22.groupby("quarter")[["total_jail_pretrial", "total_jail_pretrial_rate"]].apply(lambda df: 1 - df.isna().mean())
# displayed is nonmissing coverage for 2022 data by quarter
# using q2 makes sense (given reporting timing) and it appears to have the most complete data entries
# but coverage is still low to start with, for the new var im computing later (and possibly i need to also drop illogical rows
```

```python
# rows by quarter
v22 = vera[vera["year"] == "2022"].copy()
countmissing= (v22
       .groupby("quarter")
       .agg(
           rows=("fips","count"),
           nonmissing_pretrial=("total_jail_pretrial", lambda s: int(s.notna().sum())),
           nonmissing_pretrial_rate=("total_jail_pretrial_rate", lambda s: int(s.notna().sum())),
       )
       .sort_index())
print(countmissing)
```

```python
# make new variables!
# total_pop as denominator as requested

if "total_pop_15to64" in vera_2022_q2.columns:
    df = vera_2022_q2  # alias

    # preconvert once (cheaper than to_numeric inside a helper each time)
    num = df[["total_jail_pretrial", "total_pop_15to64"]].apply(
        pd.to_numeric, errors="coerce"
    )

    new_cols = pd.DataFrame({
        "pretrial_rate_100k_total_pop": (
            (num["total_jail_pretrial"] / num["total_pop_15to64"])
            .where(num["total_pop_15to64"].gt(0))
            * 1e5
        )
    }, index=df.index)

    vera_2022_q2 = pd.concat([df, new_cols], axis=1, copy=False)
else:
    # optionally skip creating the column at all instead of filling with NaN
    pass

vera_keep = vera_2022_q2.copy()
```

```python
import numpy as np
for c in ["total_jail_pretrial", "total_pop_15to64"]:
    if c in vera_keep.columns:
        vera_keep[c] = pd.to_numeric(vera_keep[c], errors="coerce")

# try applying some exclusion rules to drop extra small counties. change the numbers whenenver needed

# vera institute notes when citing a 2013 table [Bureau of Justice Statistics, Census of Jails (2013).]
# https://www.vera.org/publications/out-of-sight-growth-of-jails-rural-america#:~:text=Notes:%20These%20tables%20provide%20rankings,Because%20of%20reporting%20error%2C%20St.
# (roll down to the bottom 'Show Notes')
# "The 560 excluded counties are the 54 counties in the six states (Alaska, Connecticut, Delaware, Hawaii, Rhode Island, and Vermont) that do not have local jails because the state prison system also provides pretrial detention; the 145, mostly rural, counties that did not supply the necessary data to BJS; the 357, mostly rural, counties that either have pretrial incarceration rates in excess of 1,000 per 100,000 or more than 90 percent of the jail population held for other authorities. This latter group is excluded because these places are generally very small counties (e.g., fewer than 10,000 residents) in which the jail effectively serves as a regional jail or contract jail for the state and federal government. Because the preponderance of detainees are from out of the county, the county pretrial incarceration rate cannot be reliably estimated. There are a total of 25,025 pretrial detainees in these excluded counties, or an average of 70 per county. Because of reporting error, St. Louis County, MO, Allegheny County, PA, Davidson County, TN, Cameron County, TX, and El Paso County, TX use 2012 data from the Survey of Jails instead of 2013.
# The number of pretrial detainees, and the number of individuals held for other local, state, and federal authorities, is the single day count on December 31, 2013.
# note that their exclusion standard for pretrial rate uses 15-64 population as the denominator.
# if we are computing the rate differently i.e. use total_pop, we might just want to judge the new var on its own integrity
# instead of filtering rows by other vars (e.g. rate comuputed with 15-64 pop as the denominator)

POP_FLOOR = 1000
RATE_CAP  = 1000      # flag absurd rates (optional safety)

crit_pop_floor   = vera_keep["total_pop_15to64"] < POP_FLOOR
crit_missing_den = vera_keep["total_pop_15to64"].isna() | (vera_keep["total_pop"] == 0)
crit_logic       = vera_keep["total_jail_pretrial"] > vera_keep["total_pop"]
crit_rate_cap    = vera_keep["total_jail_pretrial_rate"] > RATE_CAP

bad_mask = crit_pop_floor | crit_missing_den | crit_logic | crit_rate_cap


# all rows to be excluded (preview a few)
excluded = vera_keep.loc[bad_mask].copy()
print("Excluded rows preview:\n", excluded.head(10))

# breakdown
def _count(mask): return int(mask.fillna(False).sum())

print()
print(f"  total_pop_15to64 < {POP_FLOOR}: {_count(crit_pop_floor)}")
print(f"  total_pop_15to64 missing/zero : {_count(crit_missing_den)}")
print(f"  pretrial > total_pop   : {_count(crit_logic)}")
print(f"  rate > {RATE_CAP}/100k : {_count(crit_rate_cap)}")
print(f"  UNIQUE rows excluded   : {_count(bad_mask)}")

# a cleaned copy
vera_clean = vera_keep.loc[~bad_mask].copy()
```
